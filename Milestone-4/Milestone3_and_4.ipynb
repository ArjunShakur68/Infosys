```python
test['predicted_new'] = test['StockCode'].apply(
    lambda x: 1 if popular_items.get(x, 0) > 10 else 0
)
```

```python
import pandas as pd

data = pd.read_csv("/content/online_retail.csv", encoding="latin1")
```

```python
data = data[['CustomerID', 'StockCode']]
data = data.dropna()
data['label'] = 1
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.2, random_state=42)
```

```python
popular_items = train['StockCode'].value_counts()
```

```python
import pickle

with open("recommendation_model.pkl", "wb") as f:
    pickle.dump(popular_items, f)
```

```python
import os
os.listdir("/content")
```

```python
import pickle

with open("recommendation_model.pkl", "rb") as f:
    deployed_model = pickle.load(f)
```

```python
def recommend_product(stock_code):
    if deployed_model.get(stock_code, 0) > 10:
        return "Recommended"
    else:
        return "Not Recommended"
```

```python
recommend_product("85123A")
recommend_product("22752")
```

```python
sample_items = test['StockCode'].head(10)

for item in sample_items:
    print(item, "→", recommend_product(item))
```
